```python
# filling null values, we will use two methods, random sampling for higher null values and 
# mean/mode sampling for lower null values

def random_value_imputation(feature):
    # Drop NA and sample random values equal to the number of missing values
    random_sample = dataset[feature].dropna().sample(dataset[feature].isna().sum(), random_state=42)
    # Set index of the random sample to the index of missing values in the original dataset
    random_sample.index = dataset[dataset[feature].isnull()].index
    # Replace the NA values with randomly sampled values
    dataset.loc[dataset[feature].isnull(), feature] = random_sample

# Function for mode imputation
def impute_mode(columnName):
    # Calculate the mode of the column
    mode_value = dataset[columnName].mode()[0]
    # Fill NA values with the mode
    dataset[columnName].fillna(mode_value, inplace=True)

```

```python
# Applying the imputation methods
# Check the number of null values and apply appropriate imputation method
for column in dataset.columns:
    if dataset[column].isnull().sum() > len(dataset) * 0.3:  # Example threshold: 30% or more missing
        random_value_imputation(column)
    else:
        impute_mode(column)
```
